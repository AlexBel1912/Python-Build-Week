```python
import os
import pandas as pd
from bs4 import BeautifulSoup

# leggi il file html salvato
with open(r"C:\Users\Lenovo\OneDrive\Documenti\Python\Build week Python\Sorgente Alex\SCARPE_SPORTIVE_UNDER_ARMOUR_PAGINA_1.html", "r", encoding="utf-8") as file:
    html = file.read()

# parsing dell'html
soup = BeautifulSoup(html, 'html.parser')

# assegnazione categoria in base al nome del file html salvato
categoria = "Scarpini Sportivi Under Armour"

# inizializzazione prodotti
prodotti = []

# scraping div
div_prodotti = soup.find_all('div', class_='_0xLoFW _78xIQ- EJ4MLB f4ql6o JT3_zV')

# ciclo per estrarre i dati dei prodotti da div_prodotti
for div_prodotto in div_prodotti:
    try:
        # url prodotto
        link = div_prodotto.find('a', class_='q84f1m CKDt_l LyRfpJ JT3_zV CKDt_l _2dqvZS')
        if not link or 'href' not in link.attrs:
            continue  # salta questo prodotto se manca l'URL
        url_prodotto = link['href']
        
        # marca e nome prodotto
        marca_tag = div_prodotto.find('h3', class_='OBkCPz lystZ1 m3OCL3 HlZ_Tf ja2E95 IvrHht ZkIJC- u-kVXO')
        if not marca_tag:
            continue  # salta se manca la marca
        marca = marca_tag.text.strip().upper()

        # nome prodotto
        nome_prodotto = div_prodotto.find('h3', class_='voFjEy lystZ1 m3OCL3 HlZ_Tf ja2E95 IvrHht ZkIJC- u-kVXO').text.strip()
        
        # inizializza i prezzi
        prezzo = "N/A"
        prezzo_ordinario = "N/A"
        
        # cerca prezzo corrente (indipendentemente dal prezzo ordinario)
        # cerca prima nel formato standard
        prezzo_tag = div_prodotto.find('span', class_='voFjEy lystZ1 Sb5G3D HlZ_Tf')
        if not prezzo_tag:
            # cerca su altri formati di prezzo
            prezzo_tag = div_prodotto.find('span', class_='voFjEy lystZ1 Sb5G3D Km7l2y')
        if not prezzo_tag:
            prezzo_tag = div_prodotto.find('span', class_='voFjEy lystZ1 m3OCL3 HlZ_Tf')
        
        if prezzo_tag:
            prezzo = prezzo_tag.text.strip().replace('â‚¬', '').strip()
        
        # cerca prezzo ordinario (solo se esiste)
        # cerca la sezione che contiene "prezzo ordinario"
        for p_tag in div_prodotto.find_all('p', class_='voFjEy _2kjxJ6 m3OCL3 HlZ_Tf _0xLoFW u9KIT8 _7ckuOK vSgP6A'):
            if 'Prezzo ordinario:' in p_tag.text:
                # prendi il primo span con il prezzo dopo il testo "prezzo ordinario"
                for span in p_tag.find_all('span', class_='voFjEy goVnUa m3OCL3 Yb63TQ'):
                    if 'Prezzo ordinario:' not in span.text:
                        prezzo_ordinario = span.text.strip().replace('â‚¬', '').strip()
                        break
                # se non trovato, prova con la classe alternativa
                if prezzo_ordinario == "N/A":
                    for span in p_tag.find_all('span', class_='voFjEy goVnUa m3OCL3 Yb63TQ ZiDB59 uVxVjw'):
                        prezzo_ordinario = span.text.strip().replace('â‚¬', '').strip()
                        break

        if prezzo_ordinario == "N/A":
            prezzo_ordinario = prezzo

        # se la marca Ã¨ fra le 4 note, aggiungi i dati alla lista/dizionario
        if marca in ['UNDER ARMOUR']:
            prodotti.append({
                "categoria": categoria,
                "marca": marca.upper(),
                "nome_prodotto": nome_prodotto,
                "prezzo_effettivo": prezzo,
                "prezzo_ordinario": prezzo_ordinario, 
                "url_prodotto": url_prodotto,
            })
            
    except Exception as e:
        print(f"Errore durante l'elaborazione: {str(e)}")
        continue


# crea dataframe
df_temp = pd.DataFrame(prodotti)

df_temp.to_csv('Under Armour.csv', index= False, encoding='utf-8')

# visualizza dataframe
display(df_temp)
```

```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import json
import time

def scrape_zalando_under_armour_sizes(url):
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(5)  # Attendi che la pagina si carichi completamente
    html = driver.page_source
    driver.close()
    soup = BeautifulSoup(html, 'html.parser')

    s = soup.find("script", attrs={"id": "re-concurrent-data-hydrate"})
    if s:
        try:
            s = s.text.replace("window.__hydrationDataConsume(", "")[:-1]
            d = json.loads(s)
        except json.JSONDecodeError as e:
            print(f"Errore durante la decodifica JSON: {e}")
            return pd.DataFrame()
    else:
        print("Script con id 're-concurrent-data-hydrate' non trovato.")
        return pd.DataFrame()

    taglie_prodotti = []
    for p in d.get("graphqlCache", {}).values():
        if p.get("data") and p["data"].get("product"):
            nome = p["data"]["product"]["name"]
            taglie_disponibili = []
            if p["data"]["product"].get("simples"):
                for t in p["data"]["product"]["simples"]:
                    taglie_disponibili.append(t["size"])
            else:
                taglie_disponibili = None
            taglie_prodotti.append({"nome_prodotto": nome, "taglie": taglie_disponibili})

    return pd.DataFrame(taglie_prodotti)

# URL della pagina principale delle scarpe Under Armour da uomo su Zalando
base_url = "https://www.zalando.it/scarpe-sportive-uomo/under-armour/"

# Inizializza una lista per contenere tutti i DataFrames delle pagine
all_df_taglie = []

# Scrape della prima pagina
df_page = scrape_zalando_under_armour_sizes(base_url)
if not df_page.empty:
    all_df_taglie.append(df_page)

# Funzione per trovare l'URL della pagina successiva
def find_next_page_url(soup):
    next_page_link = soup.find('a', {'aria-label': 'Vai alla pagina successiva'})
    if next_page_link and next_page_link.get('href'):
        return "https://www.zalando.it" + next_page_link['href']
    return None

# Scrape delle pagine successive
current_url = base_url
page_num = 1
while True:
    driver = webdriver.Chrome()
    driver.get(current_url)
    time.sleep(5)
    html = driver.page_source
    driver.close()
    soup = BeautifulSoup(html, 'html.parser')

    next_page = find_next_page_url(soup)
    if next_page:
        page_num += 1
        print(f"Scraping pagina {page_num}: {next_page}")
        df_page = scrape_zalando_under_armour_sizes(next_page)
        if not df_page.empty:
            all_df_taglie.append(df_page)
        current_url = next_page
    else:
        print("Nessuna altra pagina trovata.")
        break

# Concatena tutti i DataFrames in uno solo
if all_df_taglie:
    df_taglie_totale = pd.concat(all_df_taglie, ignore_index=True)
    print("\nDataFrame df_taglie_totale (Tutte le pagine):")
    print(df_taglie_totale)
else:
    print("\nNessun dato trovato da nessuna pagina.")
```

```python
import requests
import bs4
import csv

url = input("Iserisci l'URL: ")

res = requests.get(url)

try:
    res.raise_for_status()  # Solleva un'eccezione se la richiesta ha avuto problemi
except Exception as exc:
    print(f"Si Ã¨ verificato questo problema: {exc}")

#Se la richiesta Ã¨ andata a buon fine, parsiamo la pagina HTML
html_page = bs4.BeautifulSoup(res.text, "html.parser")

#Selezioniamo tutti gli elementi con la classe 'tile-body' (modifica in base alla struttura della pagina)
elem_html = "OBkCPz lystZ1 m3OCL3 HlZ_Tf ja2E95 IvrHht ZkIJC- u-kVXO"
sel_elem = html_page.select(elem_html)

#Se ci sono elementi trovati, estraiamo i dati e li scriviamo nel CSV
if sel_elem:
    with open("prodotti.csv", mode="w", newline="", encoding="utf-8") as csv_file:
        writer = csv.writer(csv_file)

        # Scriviamo l'intestazione del CSV
        writer.writerow(["Prodotto", "Descrizione"])

        # Cicliamo su tutti gli elementi trovati e li scriviamo nel CSV
        for i, elem in enumerate(sel_elem):
            # Estrai il testo da ciascun elemento e rimuovi spazi extra
            text_elem = elem.getText().strip()

            # Scriviamo una riga per ogni prodotto trovato
            writer.writerow([f"Prodotto {i+1}", text_elem])

    print(f"File CSV aggiornato con {len(sel_elem)} prodotti.")
else:
    print("Elemento con la classe 'tile-body' non trovato")
```
