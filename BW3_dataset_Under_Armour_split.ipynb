```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Under Armour Unito.csv", encoding='utf-8')
df
```

```python
# Dividi la colonna "nome_prodotto" in 3 parti
df[['modello', 'sport', 'colori']] = df['nome_prodotto'].str.split(r'\s*-\s*', n=2, expand=True)

# Salva il DataFrame aggiornato in un nuovo CSV
df.to_csv('Under_Armour_Suddiviso.csv', index=False, encoding='utf-8')
```

```python
print("\nSintesi delle informazioni strutturali e del contenuto del DataFrame:")
print(df.info())
```

```python
#Iteriamo ogni colonna del DataFrame
for column in df.columns:
    #Somiamo i valori nulli (i dati booleani vengono interpretati come 1 e 0)
    nan_count = df[column].isna().sum()
    #calcoliamo la percentuale
    nan_percentage = round((nan_count/df.shape[0])*100, 2)
    print(f'{column} contains {nan_count} NaN values, {nan_percentage}% of all rows.')
```

```python
# Crea un nuovo DataFrame contenente solo le righe dove la colonna 'colori' Ã¨ NaN
df_colori_nan = df[df['colori'].isna()]

# Visualizza il nuovo DataFrame
display(df_colori_nan)
```

```python
# Cambio il DataFrame originale senza le righe con i valori NaN
df.dropna(inplace=True)
df
```

```python
df.isnull().sum()
```

#### Siccome, nella collona colori e stato splitato con i valori nulli perche il nome_prodotto nelle righe selettate hanno soltanto il modello ha soltanto lo sport e il colore senza il modello, in questo caso sposto i valori dalle colonne 'modello' e 'sport' in 'sport' e 'colori'.

```python
# sposto i valori da 'sport' a 'colori' e sovvrascrivere i NaN esistenti
df_colori_nan.loc[:, 'colori'] = df_colori_nan['sport']

# sposto i valori da 'modello' a 'sport' nel nuovo DataFrame df_colori_nan
df_colori_nan.loc[:, 'sport'] = df_colori_nan['modello']

# pulisco la colonna 'modello' impostando con valori Unknown
df_colori_nan.loc[:, 'modello'] = 'Unknown'

# Visualizza il DataFrame
df_colori_nan
```

```python
# Ora unisco nuovamente i due DataFrame e lo visualizziamo nuovamente
df = pd.concat([df, df_colori_nan], ignore_index=True)

df
```

```python
df.isnull().sum()
```

```python
# elimino eventuali spazi iniziali e finali
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.strip()
        
#Verifichiamo le modifiche
print(df.head(5))
```

##### Converto le colonne con il prezzo effetivo e prezzo ordinario in float in quanto abbiamo visto che sono di tipo object

```python
# Sostituisci la virgola con il punto e poi converti in float solo i valori validi
df['prezzo_effettivo'] = df['prezzo_effettivo'].astype(str).str.replace(',', '.').apply(lambda x: float(x) if x.replace('.', '', 1).isdigit() else x)
df['prezzo_ordinario'] = df['prezzo_ordinario'].astype(str).str.replace(',', '.').apply(lambda x: float(x) if x.replace('.', '', 1).isdigit() else x)

# Verifica il nuovo dtype delle colonne
print(df[['prezzo_effettivo', 'prezzo_ordinario']].dtypes)
```
